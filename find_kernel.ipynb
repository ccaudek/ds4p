```python
import os
import json

# Directory where your notebooks are stored
notebook_dir = "./chapter_7"

# Loop through each file in the directory
for notebook_file in os.listdir(notebook_dir):
    if notebook_file.endswith(".ipynb"):
        # Construct the full path to the notebook
        file_path = os.path.join(notebook_dir, notebook_file)

        # Open and read the notebook
        with open(file_path, "r", encoding="utf-8") as f:
            notebook = json.load(f)

        # Check the kernel specification
        kernelspec = notebook.get("metadata", {}).get("kernelspec", {}).get("name", "")
        if kernelspec == "ir":
            print(f'The notebook "{notebook_file}" requires the R kernel.')
```
