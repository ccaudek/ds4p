# Ambiente virtuale

## Cosa è un Ambiente Virtuale?

Immaginate un ambiente virtuale come un'area di lavoro privata sul vostro computer dove potete installare e utilizzare le librerie Python (che sono come strumenti o componenti aggiuntivi per Python) senza influenzare il resto del computer. Questo ambiente è separato dal sistema principale, quindi ciò che fate all'interno non interferisce con altri programmi.

## Perché Usare un Ambiente Virtuale?

1. **Isolamento**: Dato che Python si aggiorna spesso e le librerie cambiano, usare un ambiente virtuale significa che potete scegliere e mantenere le versioni specifiche di Python e delle librerie che funzionano meglio per il vostro progetto.
2. **Organizzazione e Sicurezza**: Mantiene il vostro computer ordinato e sicuro, poiché le modifiche fatte all'interno dell'ambiente virtuale non influenzano il resto del sistema.
3. **Riproducibilità del Codice**: Se condividete il vostro codice con altre persone, l'ambiente virtuale aiuta a garantire che funzioni anche sui loro computer come funziona sul vostro.

## Come Creare un Ambiente Virtuale

Per creare un ambiente virtuale, faremo uso di uno strumento chiamato `conda`, il quale dovrebbe essere già presente sul vostro sistema se avete seguito correttamente le istruzioni per installare Anaconda.

## Passaggi per Gestire gli Ambienti Virtuali

Gestire gli ambienti virtuali è fondamentale per mantenere organizzate le dipendenze dei vari progetti. Ecco come procedere:

1. **Uscire dall'Ambiente Virtuale Corrente**:
   Se siete all'interno di un ambiente virtuale e desiderate uscirne, potete farlo utilizzando il comando:
   ```bash
   conda deactivate
   ```
   Se il sistema risponde indicando che non ci sono ambienti da disattivare, significa che non siete in un ambiente virtuale e potete procedere al passo successivo.

2. **Creare un Nuovo Ambiente Virtuale**:
   Per creare un nuovo ambiente virtuale, ad esempio denominato `pymc_env`, e installare una versione specifica di PyMC, potete utilizzare il comando:
   ```bash
   conda create -n pymc_env -c conda-forge "pymc==5.10.4"
   ```
   Conda vi chiederà di confermare l'installazione digitando `y`. Questo passaggio crea l'ambiente e installa la versione specificata di PyMC, insieme a tutte le dipendenze richieste.

3. **Attivare il Nuovo Ambiente**:
   Per utilizzare l'ambiente che avete appena creato, dovete attivarlo con:
   ```bash
   conda activate pymc_env
   ```

4. **Installare le Librerie Necessarie**:
   All'interno dell'ambiente, potete installare altre librerie necessarie per il vostro progetto. Ecco come installare alcune librerie comuni per l'analisi dati:
   ```bash
   conda install -c conda-forge jax numpyro bambi seaborn pingouin watermark -y 
   ```
   È importante notare che, per gli utenti Windows, `numpyro` potrebbe non essere supportato. In tal caso, è possibile installare un'alternativa come `nutpie`:
   ```bash
   conda install -c conda-forge nutpie
   ```

5. **Aggiornare PyMC e Altre Librerie**:
   Per mantenere aggiornate le librerie nel vostro ambiente, compreso PyMC, potete eseguire:
   ```bash
   conda update --all
   ```
   Questo comando verifica la disponibilità di versioni più recenti delle librerie installate e le aggiorna.

6. **Comandi Utili per Gestire Ambienti e Librerie**:
   Ecco alcuni comandi aggiuntivi che potrebbero tornarvi utili:
   - Per elencare tutti gli ambienti virtuali disponibili e vedere quello attivo:
     ```bash
     conda env list
     ```
   - Per rimuovere un ambiente virtuale, ad esempio `my_env`:
     ```bash
     conda env remove -n my_env
     ```
   - Per rimuovere una specifica libreria dall'ambiente, ad esempio `package_name`:
     ```bash
     conda remove -n nome_ambiente package_name
     ```
   Quest'ultimo comando richiede di specificare l'ambiente dal quale si desidera rimuovere la libreria, garantendo così che l'operazione sia eseguita nel contesto corretto.

Questi passaggi e comandi dovrebbero fornire una solida base per la gestione efficiente degli ambienti virtuali con Conda, consentendovi di mantenere separate e organizzate le dipendenze dei diversi progetti.

In conclusione, per configurare l'ambiente virtuale `pymc_env` e installare PyMC insieme agli altri pacchetti necessari per questo insegnamento, è possibile seguire le istruzioni seguenti a seconda del sistema operativo.

- Su Unix, eseguire i seguenti comandi:

   ```bash
   conda deactivate
   conda create -n pymc_env -c conda-forge "pymc==5.10.4"
   conda activate pymc_env
   conda install -c conda-forge jax numpyro bambi seaborn pingouin watermark -y 
   ```

- Su Windows, seguire queste istruzioni:

   ```bash
   conda deactivate
   conda create -n pymc_env -c conda-forge "pymc==5.10.4"
   conda activate pymc_env
   conda install -c conda-forge nutpie bambi seaborn pingouin watermark -y 
   ```
